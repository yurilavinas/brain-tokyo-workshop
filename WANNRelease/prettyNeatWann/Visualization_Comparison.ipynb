```python
# -- View Topology of Network
# Lighter color connections are connected to earlier layers (NOT weight strength)
import vis as nv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load stats:
"""
# Rows: [generations]
# Cols:
0: evaluations (one evaluation of one individual -- may be multiple rollouts)
1: median fitness 
2: median variance
3: median novelty 
4: best variance
5: elite variance
6: best variance
7: elite variance
9: top fitness (best fitness ever)
10: peak fitness (fitness earned with best single weight value by max individual)
6: median # of nodes in population
7: median # of connections in population]
"""
novelty = np.loadtxt('log/swingup_novelty_stats.out', delimiter=',') 
variance = np.loadtxt('log/swingup_var_stats.out', delimiter=',') 
multi = np.loadtxt('log/swingup_multi_stats.out', delimiter=',') 
wann = np.loadtxt('log/swingup_stats.out', delimiter=',') 
```

```python
# Plot fitness progress:
plt.figure(figsize=(12,4))
plt.grid(linestyle='--', linewidth=0.1)

plt.plot(novelty[:,2], label="Median variance (NS)")
plt.plot(variance[:,2], label="Median variance (Var)")
plt.plot(multi[:,2], label="Median variance (Var & NS)")
plt.plot(wann[:,2], label="Median variance (WANN)")

plt.legend()
plt.grid(linestyle=':', linewidth=2)
plt.xlabel('Generations')
plt.ylabel('Variance')


plt.title('ALl methods median variance')
```
