```python
# -- View Topology of Network
# Lighter color connections are connected to earlier layers (NOT weight strength)
import vis as nv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load stats:
"""
# Rows: [generations]
# Cols:
0: evaluations (one evaluation of one individual -- may be multiple rollouts)
1: median fitness 
2: median variance
3: median novelty 
4: best variance
5: elite variance
6: best variance
7: elite variance
9: top fitness (best fitness ever)
10: peak fitness (fitness earned with best single weight value by max individual)
6: median # of nodes in population
7: median # of connections in population]
"""
novelty = np.loadtxt('log/sswingup_ns_stats.out', delimiter=',') 
variance = np.loadtxt('log/sswingup_var_stats.out', delimiter=',') 
multi = np.loadtxt('log/sswingup_multi_stats.out', delimiter=',') 
wann = np.loadtxt('log/sswingup_mean_stats.out', delimiter=',') 
```

```python
# Plot fitness progress:
plt.figure(figsize=(12,4))
plt.grid(linestyle='--', linewidth=0.1)

plt.plot(novelty[:,2], label="Median variance (NS)")
plt.plot(variance[:,2], label="Median variance (Failure)")
plt.plot(multi[:,2], label="Median variance (Failure & NS)")
plt.plot(wann[:,2], label="Median variance (WANN)")

plt.legend()
plt.grid(linestyle=':', linewidth=2)
plt.xlabel('Generations')
plt.ylabel('Median Variance')


plt.title('ALl methods - Median Variance')
```

```python
# Plot fitness progress:
plt.figure(figsize=(12,4))
plt.grid(linestyle='--', linewidth=0.1)

plt.plot(novelty[:,4], label="Best Variance (NS)")
plt.plot(variance[:,4], label="Best variance (Failure)")
plt.plot(multi[:,4], label="Best variance (Failure & NS)")
plt.plot(wann[:,4], label="Best variance (WANN)")

plt.legend()
plt.grid(linestyle=':', linewidth=2)
plt.xlabel('Generations')
plt.ylabel('Best Variance')


plt.title('ALl methods - Best Variance')
```
