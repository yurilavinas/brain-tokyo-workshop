```python
# -- View Topology of Network
# Lighter color connections are connected to earlier layers (NOT weight strength)
import vis as nv
import numpy as np
from matplotlib import pyplot as plt
```

```python
nv.viewInd('log/var_best.out','bipedhard')
```

```python
# Load stats:
"""
# Rows: [generations]
# Cols:
0: evaluations (one evaluation of one individual -- may be multiple rollouts)
1: median fitness 
2: top variance (higher variance ever)
3: maximum fitness
4: top fitness (best fitness ever)
5: peak fitness (fitness earned with best single weight value by max individual)
6: median # of nodes in population
7: median # of connections in population]
"""
stats = np.loadtxt('log/var_stats.out', delimiter=',') 
```

```python
# Plot fitness progress:
plt.figure(figsize=(12,4))
plt.grid(linestyle='--', linewidth=0.1)

plt.plot(stats[:,2], label="Best variance")
plt.plot(stats[:,3], label="Elite Variance")
plt.plot(stats[:,4], label="Elite fitness")
plt.plot(stats[:,5], label="Best fitness")
plt.plot(stats[:,6], label="Best Peak Fitness")

plt.legend()
plt.grid(linestyle=':', linewidth=2)
plt.xlabel('Generations')
plt.ylabel('Reward')


plt.title('Variance method - Bipedal')

axes = plt.gca()
axes.set_ylim([0,280])
axes.set_xlim([-5,1150])
```
